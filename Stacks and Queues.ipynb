```python
# 3.2

class MultiStack:

    def __init__(self, stacksize):
        self.array = [0] * (stacksize)
        self.size = 0
        self.stacksize = stacksize
        self.minvals = [99999] * (stacksize)

    def Push(self, item):
        if self.IsFull():
            raise Exception('Stack is full')
        self.size += 1
        if self.IsEmpty():
            self.minvals[self.size - 1] = item
        else:
            self.minvals[self.size - 1] = min(item, self.minvals[self.size - 2])
        self.array[self.size - 1] = item

    def Pop(self):
        if self.IsEmpty():
            raise Exception('Stack is empty')
        value = self.array[self.size - 1]
        self.array[self.size - 1] = 0
        self.size -= 1
        return value

    def Peek(self):
        if self.IsEmpty():
            raise Exception('Stack is empty')
        return self.array[self.size - 1]

    def Min(self):
        return self.minvals[self.size - 1]

    def IsEmpty(self):
        return self.size == 0

    def IsFull(self):
        return self.size == self.stacksize



def StackMin():
    newstack = MultiStack(10)
    newstack.Push(5)
    newstack.Push(6)
    newstack.Push(2)
    newstack.Push(7)
    newstack.Push(14)
    newstack.Push(3)
    print (newstack.Min())
    newstack.Push(1)
    newstack.Push(4)
    newstack.Push(44)
    newstack.Push(2)
    print (newstack.Min())
    newstack.Pop()
    print (newstack.minvals)
    print (newstack.array)

StackMin()
```

```python
#3.5
class Stack():
    
    def __init__(self):
        self.stack = []
        
    def push(self, item):
        self.stack.append(item)
        
    def top(self):
        return self.stack[len(self.stack) - 1]
    
    def isEmpty(self):
        return len(self.stack) == 0
    
    def pop(self):
        if self.isEmpty() == False:
            return self.stack.pop()
        
    def printStack(self):
        for i in self.stack:
            print(i, end = " ")
        print()
            
    def sort(self):
        
        tempStack = Stack()
        while (self.isEmpty() == False):
            
            tmp = self.top()
            self.pop()
            
            while(tempStack.isEmpty() == False and tempStack.top() > tmp):
                self.push(tempStack.top())
                tempStack.pop()

            tempStack.push(tmp)
            
          
     
        return tempStack
    
    
    
stack = Stack()
stack.push(34) 
stack.push(3) 
stack.push(31) 
stack.push(98) 
stack.push(92) 
stack.push(23)
stack.printStack()
 
print("Sorted numbers are: ")
sortedList = stack.sort()
sortedList.printStack()          
```
